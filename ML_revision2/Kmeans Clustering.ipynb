```python
from sklearn.datasets import make_blobs
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
```

```python
X, y = make_blobs(n_samples=100, n_features=5, centers=5, random_state=80)
plt.scatter(X[:,0], X[:,1])
```

```python
X[:6]
```

```python
y[:6]
```

```python
kmeans = KMeans(n_clusters=5)
kmeans = kmeans.fit(X)
labels = kmeans.predict(X)
C= kmeans.cluster_centers_
```

```python
plt.scatter(X[:,0], X[:,1], c=y) # plot true clusters
plt.scatter(C[:,0], C[:,1], marker='*', s=1000, )
plt.title('K means clustering')
```

```python
plt.scatter(X[:,0], X[:,1], c=labels) # plot true clusters
plt.scatter(C[:,0], C[:,1], marker='*', s=1000, )
plt.title('K means clustering')
```

```python
from sklearn import datasets
iris = datasets.load_iris()
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
```

```python
KNeighborsClassifier?
```

```python
x = iris['data']
y = iris['target']
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y, random_state=25)
```

```python
knn = KNeighborsClassifier()
knn.fit(x_train, y_train)
y_pred = knn.predict(x_test)
knn.score(x_test, y_test)
print(knn.score(x_test, y_test))
misclassified = (y_test != y_pred).sum()
print(misclassified)
```

```python
knn.predict([[5.0,3.6,1.3,0.25]])
```

```python
for i in range(5,20,1):
    knn = KNeighborsClassifier(n_neighbors=i)
    knn.fit(x_train, y_train)
    y_pred = knn.predict(x_test)
    knn.score(x_test, y_test)
    print(f'----{i}--Nearest Nieghbours---', '\t',knn.score(x_test, y_test), '----', misclassified)
    print()
    misclassified = (y_test != y_pred).sum()
```
